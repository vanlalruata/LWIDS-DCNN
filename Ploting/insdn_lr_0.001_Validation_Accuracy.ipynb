```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# File paths for the datasets
file_paths = ['C:/Users/Maruata/Desktop/IO-DCNN/lr_001_adam_insdn.csv', 
              'C:/Users/Maruata/Desktop/IO-DCNN/lr_001_adagrad_insdn.csv', 
              'C:/Users/Maruata/Desktop/IO-DCNN/lr_001_adamax_insdn.csv', 
              'C:/Users/Maruata/Desktop/IO-DCNN/lr_001_nadam_insdn.csv',
              'C:/Users/Maruata/Desktop/IO-DCNN/lr_001_sgd_insdn.csv']
```

```python
# Colors and labels for each dataset
colors = ['b', 'g', 'r', 'c', 'k']
labels = ['Adam', 'Adagrad', 'Adamax', 'Nadam', 'SGD']
```

```python
# Initialize the plot
plt.figure(figsize=(12, 8))

# Iterate over each dataset
for i, file_path in enumerate(file_paths):
    # Load dataset
    data = pd.read_csv(file_path)
    
    # Plot Loss
    #plt.plot(range(1, 11), data['Loss'], color=colors[i], linestyle='-', label=f'{labels[i]} Loss')
    # Plot Accuracy
    #plt.plot(range(1, 11), data['Accuracy'], color=colors[i], linestyle='--', label=f'{labels[i]} Accuracy')
    # Plot Validation Loss
    #plt.plot(range(1, 11), data['Val_Loss'], color=colors[i], linestyle='-.', label=f'{labels[i]} Validation Loss')
    # Plot Validation Accuracy
    plt.plot(range(1, 11), data['Val_Accuracy'], color=colors[i], linestyle='-', label=f'{labels[i]} Validation Accuracy')


# Set x-axis limits and ticks for epochs 1 to 10
plt.xlim(1, 10)
plt.xticks(range(1, 11), fontsize=18)

# Add labels and title
plt.xlabel('Epochs', fontsize=16)
plt.ylabel('Metrics', fontsize=16)
plt.title('InSDN Validation Metrics Across Different Optimizer', fontsize=16)
plt.legend(loc='best', fontsize='large')
plt.grid(True)

# Save the graph as EPS format
output_filename = 'insdn_lr_001_validation_accuracy_metrics.eps'
plt.tight_layout()
plt.savefig(output_filename, format='eps', dpi=300, transparent=True)


# Show the plot
plt.show()
```
