```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#Adam
```

```python
confMat = [[  364,     0,     0,     0,     0,     0,     0,     0],
       [    0,    42,     0,     0,     0,     0,     0,     0],
       [    0,     0, 30488,     1,     9,     0,     0,     0],
       [    0,     0,     0, 13424,     0,     2,     0,     0],
       [    1,     0,     0,     1, 17051,     0,     0,     0],
       [    0,     0,     0,    14,     5, 24520,     0,     0],
       [    0,     0,     0,     1,     1,     0,     1,     0],
       [    0,     7,     0,     1,     0,     0,     0,    40]]
```

```python
# Labels
labels = ['DDoS', 'Probe', 'Normal', 'DoS', 'BFA', 'Web-Attack', 'BOTNET', 'U2R']
```

```python
# Normalize the confusion matrix for better visualization
confMat_normalized = np.array(confMat) / np.sum(confMat, axis=1, keepdims=True)

# Plot the confusion matrix
plt.figure(figsize=(12, 10))
sns.heatmap(confMat_normalized, annot=False, fmt=".2f", cmap="Blues", xticklabels=labels, yticklabels=labels)

# Add labels, title, and adjust layout
plt.xlabel("Predicted Labels", fontsize=16)
plt.ylabel("True Labels", fontsize=16)
plt.title("Confusion Matrix", fontsize=18)
plt.xticks(rotation=45, fontsize=12, ha="right")
plt.yticks(fontsize=12)

# Save as EPS file
output_filename = "insdn_lr_001_adam_confusion_matrix.eps"
plt.tight_layout()
plt.savefig(output_filename, format="eps", dpi=300)

# Show the plot (optional)
plt.show()
```
